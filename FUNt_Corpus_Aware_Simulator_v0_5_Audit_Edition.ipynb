```python
# ============================================================
# FUNt Corpus-Aware Simulator v0.5 - Audit Edition
# Locked Reference Build
# ============================================================

import numpy as np
import hashlib
import json
from datetime import datetime

np.set_printoptions(suppress=True, precision=6)

# ===================== CONFIG =====================

NARRATIVE_MODE = True
STEPS_PER_STAGE = 3

INITIAL_FIELD = [1.0, 1.15, 1.35, 1.55, 1.70, 1.60, 1.40, 1.20, 1.05, 1.0]

# ===================== PARAMETER LEDGER =====================

ledger = {
    "version": "v0.5 - Audit Edition",
    "status": "Locked Reference Build",
    "narrative_mode": NARRATIVE_MODE,
    "steps_per_stage": STEPS_PER_STAGE,
    "initial_field": INITIAL_FIELD,
    "operators": ["H0", "Log-Psi", "Transport_2sqrt", "h3pi", "hHRT"],
    "classification": {
        "H0": "Stable",
        "Log-Psi": "Stable",
        "Transport_2sqrt": "Stable",
        "h3pi_operator": "Stable",
        "h3pi_transition_detection": "Exploratory Application",
        "hHRT": "Stable"
    },
    "limits": {
        "transition_threshold": 0.15,
        "transport_strength": 0.08,
        "relaxation_rate": 0.05
    }
}

run_hash = hashlib.sha256(
    json.dumps(ledger, sort_keys=True).encode("utf-8")
).hexdigest()

print("=" * 82)
print("FUNt Corpus-Aware Simulator v0.5 - Audit Edition")
print("Status: Locked Reference Build")
print(f"Run Timestamp: {datetime.utcnow().isoformat() } UTC")
print(f"Run Hash: {run_hash}")
print(f"Mode: {'Narrative Mode' if NARRATIVE_MODE else 'Pure Operator Mode'}")
print("=" * 82)

# ===================== CORE OPERATORS =====================

def H0(field):
    correction = np.mean(field)
    return field - correction, abs(correction)

def Log_Psi(field):
    return np.gradient(field)

def Transport_2sqrt(field, strength=0.08):
    grad = Log_Psi(field)
    delta = np.tanh(grad) * strength
    return field + delta, float(np.mean(np.abs(delta)))

def h3pi_transition_detection(field, threshold=0.15):
    return np.abs(Log_Psi(field)) > threshold

def hHRT(field, rate=0.05):
    deviation = field - np.mean(field)
    delta = rate * deviation
    return field - delta, float(np.mean(np.abs(delta)))

# ===================== STAGES =====================

STAGES = [
    ("Stage 1: Primordial Hydrogen Coherence", "H0 baseline established"),
    ("Stage 2: Electron Magnetic Alignment", "Recurrence begins"),
    ("Stage 3: Lattice Formation", "Constrained transport organizes structure"),
    ("Stage 4: Gravitational Strain", "Transition regions emerge"),
    ("Stage 5: Unified Wave-Web", "Relaxation toward coherent state")
]

# ===================== SIMULATION =====================

def run_simulation(initial_field):
    field = np.array(initial_field, dtype=float)
    influence_log = []
    total_transitions = 0

    for stage_name, stage_note in STAGES:
        if NARRATIVE_MODE:
            print(f"\n{stage_name}")
            print(f"  Interpretive Note: {stage_note}")

        for step in range(1, STEPS_PER_STAGE + 1):
            field, h0_correction = H0(field)
            field, transport_delta = Transport_2sqrt(field)

            transitions = h3pi_transition_detection(field)
            transition_count = int(np.sum(transitions))
            total_transitions += transition_count

            field, relaxation_delta = hHRT(field)

            influence_log.append({
                "stage": stage_name,
                "step": step,
                "H0_correction": round(float(h0_correction), 6),
                "Transport_2sqrt_delta": round(float(transport_delta), 6),
                "h3pi_transition_count": transition_count,
                "hHRT_relaxation_delta": round(float(relaxation_delta), 6)
            })

            if NARRATIVE_MODE and transition_count > 0:
                print(f"    Step {step}: {transition_count} transition(s) detected")

    return field, influence_log, total_transitions

# ===================== AUDIT =====================

def audit_run(final_field, total_transitions):
    audit_flags = []

    final_mean = float(np.mean(final_field))
    final_std = float(np.std(final_field))
    total_possible = len(STAGES) * STEPS_PER_STAGE * len(final_field)
    transition_ratio = total_transitions / total_possible

    if final_std > 0.8:
        audit_flags.append("HIGH VARIANCE")

    if transition_ratio > 0.35:
        audit_flags.append("EXCESSIVE TRANSITIONS")

    if abs(final_mean) > 0.3:
        audit_flags.append("POOR RELAXATION TO H0")

    return final_mean, final_std, transition_ratio, audit_flags

# ===================== RUN =====================

final_field, influence_log, total_transitions = run_simulation(INITIAL_FIELD)
mean, std, transition_ratio, audit_flags = audit_run(final_field, total_transitions)

print("\n" + "=" * 82)
print("FINAL STATE")
print(f"Final Field: {np.round(final_field, 5)}")
print(f"Mean: {mean:.5f}")
print(f"Std: {std:.5f}")
print(f"Transition Ratio: {transition_ratio:.3f}")

print("\nAUDIT FLAGS:")
if audit_flags:
    for flag in audit_flags:
        print(f"  WARNING: {flag}")
else:
    print("  None triggered")

print("\nOPERATOR INFLUENCE SUMMARY")
for entry in influence_log[:5]:
    print(f"  {entry}")

# ===================== LIMITATIONS =====================

print("\n" + "=" * 82)
print("LIMITATIONS STATEMENT")
print("""
This simulator is an exploratory operator sandbox only.

- It is not a validation engine.
- It is not a cosmology simulator.
- It is not evidence that FUNt is correct or physically real.
- Narrative descriptions are interpretive only.
- Only the five core operators evolve the field.
- Users, especially students, may perceive greater theoretical maturity
  than the current mathematical support justifies.

Any scientific conclusion requires independent institutional testing.
""")
print("=" * 82)
print(f"Run Hash: {run_hash}")
print("v0.5 complete")
```
